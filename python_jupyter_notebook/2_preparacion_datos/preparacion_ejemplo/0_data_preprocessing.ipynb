```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. LOAD DATA

# import data
datos1 = pd.read_csv("Datos1.csv");
datos2 = pd.read_excel("Datos2.xlsx");
```

```python
# debug imported data
datos1 # framedata 1
```

```python
# debug imported data
datos2 # framedata 2
```

```python
# 2. INTEGRATE DATA

# make a sql join between framedata 1 and framedata 2
# read sql comparision here: https://pandas.pydata.org/pandas-docs/stable/comparison_with_sql.html
data = pd.merge(datos1, datos2, on='Id')
```

```python
# 3. DESCRIPTIVE ANALISIS

# data analisis functions http://pythonhow.com/pandas-data-analysis-functions/

print("Entregas Completas\n- mean: ")
data['Entregas_Completas'].mean()
```
