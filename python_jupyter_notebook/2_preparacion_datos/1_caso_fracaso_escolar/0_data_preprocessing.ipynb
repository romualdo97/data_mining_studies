```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. LOAD DATA

```python
# import data
datos1 = pd.read_csv("Datos1.csv");
datos2 = pd.read_excel("Datos2.xlsx");
```

```python
# debug imported data
datos1 # framedata 1
```

```python
# debug imported data
datos2 # framedata 2
```

# 2. INTEGRATE DATA

```python
# make a sql join between framedata 1 and framedata 2
# read sql comparision here: https://pandas.pydata.org/pandas-docs/stable/comparison_with_sql.html
data = pd.merge(datos1, datos2, on='Id')
data
```

# 3. DESCRIPTIVE ANALISIS

```python
# data analisis functions http://pythonhow.com/pandas-data-analysis-functions/

# dataframe for describe numeric values
d = {'Asistencia': data['Asistencia'], 'Entregas_Completas': data['Entregas_Completas']}
df4descriveNumerics = pd.DataFrame(data=d)
df4descriveNumerics
```

```python
df4descriveNumerics.describe()
```

```python
# dataframe for describe categorical values, read more here 
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.describe.html
d = {'Estrato': data['Estrato'].astype('category'), 
    'Sexo': data['Sexo'].astype('category'),
    'Enfermedad': data['Enfermedad'].astype('category'),
    'Colegio_U': data['Colegio_U'].astype('category'),
    'Activo_Web': data['Activo_Web'].astype('category'),
    'Trabaja': data['Trabaja'].astype('category'),
    'Examen_Final': data['Examen_Final'].astype('category')}
df4descriveCategorical = pd.DataFrame(data=d)
df4descriveCategorical
```

```python
df4descriveCategorical.describe()
```
